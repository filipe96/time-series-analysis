```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prefixspan import PrefixSpan
import plotly.plotly as py
```

```python
filepath = "a_ereignis_02.csv"
delimiter = "|"
```

```python
def readFile(file, delim):
    dataframe = pd.read_csv(filepath_or_buffer=file, delimiter=delim)
    return dataframe
```

```python
def filterDataFrame(df):
    table = df[['$COLUMNS$MASCH_NR', 'BEGIN_ZEIT', 'ENDE_ZEIT', 'BEGIN_DAT', 'STOERTXT_NR']]
    table.columns = ['MashineID', 'StartTime', 'FinishTime', 'Date', 'MashineState']
    table = table[['MashineID','StartTime', 'FinishTime', 'Date','MashineState']]
    table = table[table['MashineID'] == 'BFD1ERF1']
    table = table[table['Date'] == '08/09/2013']
    table =  table[table['MashineState'] == 2] 
    return table.sort_values(by=['StartTime'])
    
```

```python
data = readFile(filepath, delimiter)
filteredDataFrame = filterDataFrame(data)
print(filteredDataFrame)
```

One Day has 86400 Seconds. A Day in the Date Set Starts at Second 0 and Ends at second 86400. The concluse is, that a time_series for a mashine has 86400. One Element for every second.

```python
#Generate an Array with 86400 seconds default 0
keys = range(86400)
machine_status_time_line = {}
for i in keys:
        machine_status_time_line[i] = 0        
```

```python
for index, row in filteredDataFrame.iterrows():
    for x in range(row['StartTime'], row['FinishTime'] + 1, 1):
        machine_status_time_line[x] = 1
```
